# 导入pandas和numpy(5分）

```python
import pandas as pd
import numpy as np
```

# 导入外部数据，数据(5分）

```python
data=pd.read_excel("~/002230.xlsx")
```

# 数据框拆分(20分）

```python
##我只提供了一个数据文件,导入后，随意命名
# 要求1：
# 请把这一个DataFrame(DF)按不同的ticker拆分成三个不同的DataFrame(DF),
# 关联命令
# newDF=DF.iloc[:,:]
```

```python
kdxf=data.iloc[:12]
sasd=data.iloc[12:24]
hsas=data.iloc[24:]
```

# 数据框调整索引(10分）

```python
# 要求2：
# columns和拆分前保持一致，记得重置index或直接把日期换成index
# 关联命令
# DF.set_index("XXX",inplace=True)
# DF.reset_index(inplace=True)
```

```python
kdxf.set_index('date',inplace=True)
sasd.set_index('date',inplace=True)
hsas.set_index('date',inplace=True)
```

# 数据框重命名和拼接（30分)

```python
# 要求3：
# 三个DF只拼接收盘价这一列到一个新的DF，并展示前五行
# 关联命令
# pd.concat([newDF1.iloc[:,:],newDF.iloc[:,:],newDF.iloc[:,:]],axis=1)
# DF.rename(columns={"XXX":"XXX"},inplace=True)
# DF.head()
```

```python
kdxf.rename(columns={"close":"close_kdxf"},inplace=True)
sasd.rename(columns={"close":"close_sasd"},inplace=True)
hsas.rename(columns={"close":"close_hsas"},inplace=True)
concat=pd.concat([kdxf["close_kdxf"],sasd["close_sasd"],hsas["close_hsas"]],axis=1)
concat.head()
```

# 计算对数收益率（10分）

```python
# 要求4：
# 三个股票都要计算以12月1日为基准的收益率。
# 关联命令
# dataconcat/dataconcat.iloc[0]
# (DF/DF.iloc[0]).head()
# np.log(DF/DF.iloc[0]).head()#收益率，lnx
```

```python
ans=np.log(concat/concat.iloc[0])
ans.head()
```

# 画出对数收益率（10分）

```python
# 要求5：
# 关联命令
# np.log(DF/DF.iloc[0]).plot(figsize=(20,10),grid=True)##将个股的每日股价按首个交易日进行归1处理并可视化
```

```python
np.log(concat/concat.iloc[0]).plot(figsize=(20,10),grid=True)
```
